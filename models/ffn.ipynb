### Sample training of a FFN net

Use the following parameters to train a Feed Forward Network. For now, loss needs to be a preexisting loss from here:https://www.tensorflow.org/api_docs/python/tf/keras/losses

The internal shapes defines the architecture of the network. For example, [32, 32] means a network with two layers, each with 32 nodes. Use [ ] for a linear regression.

After training, the trained model will be saved on ./models/\< model_name 
\> and can be loaded 


```python
loss = 'MSE'
learning_rate = 0.001
epochs = 50 # How many training steps we want
shape = [32, 32]
model_name = ffn.get_name(learning_rate, shape)
verbose = 1 # Prints training progress
```

```python
model, test_loss = ffn.announcement_ffn(data = data, model_name=model_name, \
    loss_=loss, learning_rate_=learning_rate, epochs=epochs, internal_shapes=shape, verbose=verbose)
```
